```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

cols = [69, 101, 118]
catalog = np.loadtxt('/data/graciana/data.mfmtk', delimiter=',', dtype=float, usecols=cols).T

font = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'normal',
        'size'   : 18,
        }

eliptica = catalog.T[np.where(catalog[0] == 1)].T
classica = catalog.T[np.where(catalog[0] == 2)].T
pseudo = catalog.T[np.where(catalog[0] == 3)].T
```

# Gráfico sózinho

```python
%matplotlib notebook
plt.plot(eliptica[1], eliptica[2], 'og', label=r"$\rm El\'iptica$")
plt.plot(classica[1], classica[2], 'sb', label=r'$\rm Cl\'assica$')
plt.plot(pseudo[1], pseudo[2], 'vr', ms=6.5, label=r'$\rm Pseudo$')
plt.xlabel(r'$\rm C_1$', fontsize=22)
plt.ylabel(r'$\rm H$', fontsize=22)
plt.ylim([0.205, 0.79])
plt.xlim([0.01, 1.1])
#plt.title(r'$\rm T\'itulo$')
plt.legend(loc=1)
plt.tick_params(labelsize=20)
#x = np.linspace(0.2, 1, 1000)
#y = np.linspace(0.2, 0.4, 1000)
#plt.plot(x, y*1, 'k')
plt.savefig('H_C')
```

#  Gráficos em grupo

```python
cols = [69, 121, 27, 4, 55]
catalog = np.loadtxt('/data/graciana/data.mfmtk', delimiter=',', dtype=float, usecols=cols).T

font = {'family' : 'serif',
        'color'  : 'black',
        'weight' : 'normal',
        'size'   : 18,
        }

eliptica = catalog.T[np.where(catalog[0] == 1)].T
classica = catalog.T[np.where(catalog[0] == 2)].T
pseudo = catalog.T[np.where(catalog[0] == 3)].T
```

```python
%matplotlib notebook
f, axes = plt.subplots(3, 1, figsize=(9, 15))



for i, ax in enumerate(axes.flat):
    if (i < 2):
        ax.set_xticks([])
    

plt.subplots_adjust(hspace=0, wspace=0)

axs = axes.flat



ax = axs[0]

ax.set_ylabel(r'$\rm dT_r$', fontsize=20)
ax.plot(eliptica[1], eliptica[2], 'og')
ax.plot(classica[1], classica[2], 'sb')
ax.plot(pseudo[1], pseudo[2], 'vr')
ax.set_ylim([0.15, 1])

ax = axs[1]

ax.set_ylabel(r'$\rm I0d_r$', fontsize=20)
ax.plot(eliptica[1], eliptica[3], 'og')
ax.plot(classica[1], classica[3], 'sb')
ax.plot(pseudo[1], pseudo[3], 'vr')
ax.set_ylim([18, 23])
ax.set_yticks([22,21, 20, 19])

ax = axs[2]

ax.set_ylabel(r'$\rm Mrd$', fontsize=20)
ax.set_xlabel(r'$\rm M_i$', fontsize=20)
ax.plot(eliptica[1], eliptica[4], 'og')
ax.plot(classica[1], classica[4], 'sb')
ax.plot(pseudo[1], pseudo[4], 'vr')
ax.set_ylim([-22.5, -18])
ax.set_yticks([-22, -20, -19])

plt.savefig('/data/mi_dtr_i0dr_mrd.png', dpi=300)
```
